# Cropping an image by removing 0-pixels and centering by center of mass

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import imageio as iio
```

## Using `zero_crop_and_center` Functions

```python
from zero_crop_and_center import find_zero_bounding_box
```

```python
img = iio.imread('some_mnist/101.png')
top, bottom, left, right = find_zero_bounding_box(img)
plt.imshow(img[top:bottom, left:right], cmap='binary')
plt.show()

img = iio.imread('some_mnist/102.png')
top, bottom, left, right = find_zero_bounding_box(img)
plt.imshow(img[top:bottom, left:right], cmap='binary')
plt.show()

img = iio.imread('some_mnist/103.png')
top, bottom, left, right = find_zero_bounding_box(img)
plt.imshow(img[top:bottom, left:right], cmap='binary')
plt.show()
```

```python
from zero_crop_and_center import center_in_image
```

```python
img = iio.imread('some_mnist/103.png')
top, bottom, left, right = find_zero_bounding_box(img)


res = center_in_image(image_array=img[top:bottom, left:right],
                      output_size=(28, 28))

plt.imshow(res, cmap='binary')
plt.show()
plt.imshow(iio.imread('some_mnist/103.png'))
plt.show()
```
